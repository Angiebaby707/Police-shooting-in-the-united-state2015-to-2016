- A bar chart (or bar graph) is a type of data visualization that uses rectangular bars to represent data values. 
Each bar's length or height is proportional to the value it represents, allowing for easy comparison betweencategories.

- Key Features of a Bar Chart:
Bars: Represent individual categories (e.g., countries) and their values (e.g., population).

- Axes:
The horizontal axis (x-axis) typically represents categories (e.g., country names).

- The vertical axis (y-axis) represents numerical values (e.g., population size).

- Bar Height/Length: Corresponds to the value of the category (the taller or longer the bar, the higherthevalue).

- Explanation 
bars: Holds the bar objects created by plt.bar().

- plt.text(): Adds text labels on top of each bar:

- bar.get_x() + bar.get_width()/2. centers the label on the bar.

- height * 1.01 positions the label slightly above the bar.

- f'{height:,}' formats the population with commas for better readability.

- ha='center' and va='bottom': Align the text horizontally in the center and vertically at the bottomofthetext.

```python
# Creating Bars
import matplotlib.pyplot as plt
# Sample data : Country names and their populations
countries = ['china','india','USA','indonesia','pakistan']
population = [1444216107, 1393409038, 332915073, 276361783, 225199937]

plt.figure(figsize=(10,6))
bars = plt.bar(countries,population,color = 'red')

# Adding title and labels
plt.title('Population of Countries', fontsize=16)
plt.xlabel('Countries')
plt.ylabel('Population')

# Adding labels on top of each bar
for bar in bars:
    height=bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height * 1.01, # Adjust position
             f'{height:,}', ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# Country names and their populations
countries = ['china','india','USA','indonesia','pakistan']
population = [1444216107, 1393409038, 332915073, 276361783, 225199937]

# Create a horizontal bar chart
plt.figure(figsize=(10,6))
bars = plt.barh(countries,population,color = 'grey')

# Adding title and labels
plt.title('Population of Countries', fontsize=16)
plt.xlabel('Countries', fontsize=12)
plt.ylabel('Population', fontsize=12)

# Adding labels to the right side of each bar
for bar in bars:
    width=bar.get_width()
    plt.text(width * 1.01, bar.get_y() + bar.get_height()/2, # Adjust position
             f'{width:,}', ha='left', va='center', fontsize=10)
```

```python
# How many rows and columns are in the dataset?
# What are the column names, and what data types do they have?
# Missing Data:

# Are there any missing values in the dataset? If so, which columns have the mostmissingdata?

import pandas as pd
police = pd.read_csv ('police.csv')
police
```

```python
# what is the most common race of the people involved in the police shooting
most_common_race = police['race'].value_counts().idxmax()
race_counts = police['race'].value_counts()
most_common_race,race_counts

# Create a bar chart for race counts
plt.figure(figsize=(20,12))
plt.bar(race_counts.index,race_counts.values,color='green')
```

```python
# count of gender
gender_count = police.count
gender_count
```

```python
# Race and Gender Analysis
# What is the breakdown of incidents by race and gender
race_gender_breakdown = police.groupby(['race','gender']).size()
race_gender_breakdown
```

```python
#Location based analysis
# what state have the highest number of shooting
state_counts = police['state'].value_counts()
state_counts
```

```python
# Ensure the 'date'column is in datetime format
police['date'] = pd.to_datetime(police['date'],errors = 'coerce')
```

```python
# Time based Analysis
# Which month do most police shootings occur
incidents_per_month = police['date'].dt.month.value_counts()
incidents_per_month
```

```python
# What is the most common race of the people involved in the police shootings
most_common_race = police['race'].value_counts().idxmax()
race_counts = police['race'].value_counts()
most_common_race,race_counts
```

```python
# How many incidents occured each year
incidents_per_year = police['date'].dt.year.value_counts()
incidents_per_year
```
